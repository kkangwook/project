```python
def comma_remover(x):
    return int(x.replace(',',''))
```

```python
data['like']=data['like'].apply(comma_remover)
```

```python
data
```

```python
import re
```

```python
a=re.compile(r'\w+')
b=a.match(df['genre'][0])
b.group()
```

```python
def one_genre(x):
    a=re.compile(r'[,].*')
    return re.sub(a,'',x)
```

```python
data['genre']=data['genre'].apply(one_genre)
```

```python
data
```

```python
data.drop(index=data[(data['genre']=='국내영화')|(data['genre']=='국내드라마')|(data['genre']=='-')|(data['genre']=='POP')|(data['genre']=='일렉트로니카')].index,inplace=True)
data
```

```python
data['genre'].unique()
```

```python
def camelot_sin(x):
    angle = 2 * np.pi * (x - 1) / 12
    return np.sin(angle)
```

```python
def camelot_cos(x):
    angle = 2 * np.pi * (x - 1) / 12
    return np.cos(angle)
```

```python
def number(x):
    return int(x[:-1])
```

```python
data['camelot_sin']=np.round(data['camelot'].apply(number).apply(camelot_sin),2)
```

```python
data['camelot_cos']=np.round(data['camelot'].apply(number).apply(camelot_cos),2)
```

```python
data
```

```python
def get_str(x):
    return 0 if x[-1]=='A' else 1
```

```python
data['camelot_AB']=data['camelot'].apply(get_str)
```

```python
data
```

```python
data.drop(columns='camelot',inplace=True)
```

```python
final_data=pd.get_dummies(data,columns=['genre'],drop_first=True,dtype='int')
final_data
```

```python
final_data['year']=data['year_y']
final_data
```

```python
final_data.columns
```
